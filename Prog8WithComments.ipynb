```
import pandas as pd
import numpy as np
```

```
import matplotlib.pyplot as plt
```

```
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
```

```
dp=pd.read_csv("/content/drive/MyDrive/datasets/8-Kmeans_EM.csv")
dp
```

```
# since it has 2 rows take it as v1 and v2
f1=np.array(dp['V1'])
```

```
f2=np.array(dp['V2'])
```

```
x=np.array(list(zip(f1,f2)))
```

```
plt.scatter(f1,f2,c="black",s=10)
```

```
# apply k means 

km=KMeans(n_clusters=3)
km=km.fit(x)
label=km.predict(x)

plt.scatter(x[:,0],x[:,1],c=label,s=10)
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],marker="*",s=10)
```

```
# apply gaussian mixture

gm=GaussianMixture(n_components=3)
gm.fit(x)
label=gm.predict(x)

print("The probability is ",gm.predict_proba(x))
print("The mean is ",gm.means_)
print("the covraince is ",gm.covariances_)

```
