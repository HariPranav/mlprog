```
# no dataset , x_train and y_train 30 values, then convert to 3d
#define a=0.5
# x_test get i/10 and y_test empty
# apply formula - wta =np.exp(np.sum-(x_train-x_test)**2)/(2*a)**2 => epow(x-x0)^2/(2a^2)
# apply formula - W=diag(wta)
# apply formula - beta= X^T WX^-1 X^T WY . Apply dot product (X^TXW)^-1 .(X^TWY)
# apply formula - wp=beta*x_train
```

```


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#used to get the background with squares
plt.style.use("seaborn")
a=0.5
```

```

x_train = np.array(list(range(3, 33)) )
print (x_train)
#2D array will become 3D array
x_train=x_train[:,np.newaxis]
print (x_train)
#y_train must have 30 values same as xtrain
y_train = np.array([1,2,1,2,1,1,3,4,5,4,5,6,5,6,7,8,9,10,11,11,12,11,11,10,12,11,11,10,9,8])
print (y_train)
```

```
#print the plot
plt.scatter(x_train, y_train )
```

```
# go from 0 to 39.9 as the loop starts from i/10 to 400
x_test = np.array([i/10. for i in range(400)])
print (x_test)
#convert 2d to 3d and then print
x_test=x_test[:,np.newaxis]
print (x_test)
#use an empty array y for the predicted value
y_test = []
```

```
for r in range(len(x_test)):
 wts=np.exp(-np.sum((x_train-x_test[r])**2,axis=1)/(2*a)**2)
# takes the diagonal elements in the array and uses it for calculation
 W=np.diag(wts)
 factor1 = np.linalg.inv(x_train.T.dot(W).dot(x_train)) # find inverse of (X.T*W*X)
 parameters=factor1.dot(x_train.T).dot(W).dot(y_train) # final values of theta
 prediction=x_test[r].dot(parameters)
 y_test.append(prediction)
y_test = np.array(y_test)
plt.scatter(x_train, y_train )
plt.scatter(x_test, y_test,color='red')


plt.xlabel('x')
plt.ylabel('Y')
plt.show()
```
