```python
import pandas as pd
import numpy as np
#we need gaussian mixture for 8th one as it contains em algo
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
#need to find the mean varience 
import sklearn.metrics as sm
import matplotlib.pyplot as plt

```

```python
data=pd.read_csv("prog8-Kmeans_EM.csv")
#split the data into two parts
f1=data['V1'].values
f2=data['V2'].values
#use zip function to map similar indices to differnt clusters
x=np.array(list(zip(f1,f2)))
# s denotes the pixels
plt.scatter(f1,f2, c='black',s=7)
plt.title("Dtaset")
```

```python
k=3
kmeans=KMeans(n_clusters=k)
kmeans=kmeans.fit(x)
labels=kmeans.predict(x)
#get the centroid values
centroids=kmeans.cluster_centers_
print("CENTROIDS")
print(centroids)
plt.title("KMeans")
#get the two columns of the arrray and map to rainbow color
plt.scatter(x[:,0],x[:,1],c=labels, cmap='rainbow',s=100)
#use start to mark the center of the clusters using the centroids
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='*',color='black')
```

```python
k=4
kmeans=KMeans(n_clusters=k)
kmeans=kmeans.fit(x)
labels=kmeans.predict(x)
centroids=kmeans.cluster_centers_
print("CENTROIDS")
print(centroids)
plt.title("KMeans")
plt.scatter(x[:,0],x[:,1],c=labels, cmap='rainbow',s=100)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='*',color='black')
```

```python
gmm=GaussianMixture(n_components=3).fit(x)
labels=gmm.predict(x)
#viridis is a pattern for vizualization.d
plt.scatter(x[:,0],x[:,1],c=labels,s=50,cmap='viridis')
plt.show()
print("EM predictions")
probs=gmm.predict_proba(x)
print(probs)
print("Kmeans:\n",gmm.means_)
print("Covariances:\n",gmm.covariances_)
```
